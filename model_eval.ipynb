```python
def NER(name, data):
  nlp=spacy.load(name)
  df=pd.DataFrame(columns=["docId","Entity","Start","End","Tag"])
  for id,text in data.items():
    print(f"\nExpediente {id}")
    doc=nlp(text)
    displacy.render(doc, jupyter=True, style='ent')
    for ent in doc.ents:
        df.loc[len(df)]={"docId":id,"Entity":ent.text, "Start":ent.start_char, "End":ent.end_char, "Tag":ent.label_}
  return df
```

```python
def extraer_unidades(datos):
    unidades={}
    for i in range(len(datos)):
        for j in range(len(datos[i])):
            if type(datos[i].iloc[j,-1]) != float:
                etiquetas=json.loads(datos[i].iloc[j,-1])
                unidad=etiquetas.get("unit","nan")

                if unidad != "nan":
                    unidades[unidad]=unidades.get(unidad,0)+1
    return unidades
```

```python
def unidades_cantidad(df,unidades, col):
    unidad_identificada=[]
    entidad=df[col]
    encontrado=False
    df["Unidad"]=None

    for i,registro in enumerate(entidad):
        registro=registro.split()
        for token in registro:
            if max(list(map(lambda uni: fuzz.ratio(token,uni),unidades)))>88:
                df.iloc[i,-1]=token

            else:
                for u in unidades:
                    patron=re.search(rf"\d+{re.escape(u)}|{re.escape(u)}\.",token)
                    patron=  patron.group() if patron else ""#regresa el elemento que encontro, en su defecto una cadena vacia
                    if patron!="":
                        df.iloc[i,-1]=u
    return df
```

```python

#Tokenizador Biomedico
model = FastText(sentences = list(train_text.values())+list(test_text.values()), window=10, min_count= 1, min_n=2)
```

```python
punctuation = ':;,.)(][}{'
```

```python
#Devuelve lista de tokens
def tokenizer(text):
  doc = nlp(text)
  return [word.text for word in doc]

#Escapar caracteres para evitar errores a la hora de usar expresiones regulares
def replace(entity):
    for sim in ["[","]","(",")","+"]:
        if sim in entity :
            entity=entity.replace(sim,f"\{sim}")
    return entity
```

```python

#Utiliza los indices del tsv para obtener las oraciones

def get_sentences(text, e1, e2):
  '''
  entity_text = [measuredP, measuredE, quantity]
  '''
  result = None
  text_split = []
  # se obtienne las sentencias y su indice
  doc = nlp(text)
  for num, sent in enumerate(doc.sents):
    #print(sent)
    if e1 in sent.text and e2 in sent.text:
      result = sent.text
  if not result:
    #print('entre al if')
    text = text.split()
    e1 = e1.split()
    e2 = e2.split()
    try:
      for i, palabra in enumerate(text):
        aux = ''
        if palabra[-1] in punctuation:
          aniade = palabra[:-1]
          aux = aniade
          if aniade[0] in punctuation:
            aniadir = aniade[0:]
            aux = aniadir
        text_split.append(aux)

      for i, palabra in enumerate(e2):
        if palabra[-1] in punctuation:
          e2[i] = palabra[:-1]

      for i, palabra in enumerate(e1):
        if palabra[-1] in punctuation:
          e1[i] = palabra[:-1]

      if not result:
        #print('Entra try')
        e1_inicio = text_split.index(e1[0])
        e2_fin=re = text_split.index(e2[-1])
        in_between = text[e1_inicio - 1:e2_fin + 1]
        in_between = ' '.join(in_between)
        result = in_between
      if not result:
        #print('Entra error 1')
        e2_inicio = text_split.index(e2[0])
        e1_fin = text_split.index(e1[-1])
        in_between = text[e2_inicio - 1:e1_fin + 1]
        in_between = ' '.join(in_between)
        result = in_between 
    except:
      result = ' '.join(text)
  return result

def shortest_path1(sentence, e1, e2):#Busca el camino mas corto en el arbol sintactico
  sentence=sentence.lower()
  e1=e1.lower()
  e2=e2.lower()
  doc = nlp(sentence)
  entity1= tokenizer(e1)
  entity2= tokenizer(e2)
  edges=[]
  for token in doc:
    for child in token.children:
      edges.append(('{0}'.format(token.text),
                      '{0}'.format(child.text)))
  graph = nx.Graph(edges)
  path = []
  try:
    way = nx.shortest_path(graph, source=entity1[-1], target= entity2[-1])
    for node in way:
      if node not in path:
        path.append(node)
    dependences = []
    for token in doc:
      for p in path:
        if p == token.text:
          if token.head.text in path and token.text in path:
            dependences.append(token.dep_)
    sintactic = ' '.join(dependences)
  except:
    # en caso de que no encuentre un camino mas cercano regresa las dependencias
    # de la entidad 1 y de la entidad dos
    dependences1 = []
    dependences2 = []
    entity1= e1.split()
    entity2= e2.split()
    for i, palabra in enumerate(entity2):
      if palabra[-1] in punctuation:
        entity2[i] = palabra[:-1]
      
    for i, palabra in enumerate(entity1):
      if palabra[-1] in punctuation:
        entity1[i] = palabra[:-1]
        
    for token in doc:
      token_text = token.text
    if token_text[-1] in punctuation:
      token_text = token.text[:-1]
      for we1 in entity1:
        if we1 == token_text:
          dependences1.append(token.dep_)
      for we2 in entity2:
        if we2 == token_text:
          dependences2.append(token.dep_)
    sintactic = ' '.join(dependences1 + dependences2)
      
  return sintactic

def BOW (e1,e2,sentence):
  e1=replace(e1) 
  e2=replace(e2) 
  try:
    find=re.search(f"{e1}(.*){e2}",sentence) 
    e1_inicio,e2_fin=find.span() 
    in_between = sentence[e1_inicio:e2_fin] 
  except IOError:
    find=re.search(f"{e2}(.*){e1}",sentence) 
    e1_inicio,e2_fin=find.span() 
    in_between = sentence[e1_inicio:e2_fin]
  except:
    in_between = sentence
  return in_between 

def features_entity1(e1, e2,text,docId):
    words1 = e1.split()[0]
    words2 = e2.split()[0]
    sentence = get_sentences(text[docId], e1, e2)
    sintactic = shortest_path1(sentence, e1, e2)
    bow = BOW (e1,e2,sentence)
    if sentence and sintactic:
      array = [model[words1], model[words2], model[bow], model[sintactic]] 
      array = np.concatenate(array, axis = 0).reshape(-1)
      return array
```

```python
def generate_df(text, df1, df2, target1, target2):
    '''
    text: diccionario que contiene el id y el texto plano de cada documento
    df1, df2: DataFrame de la entidad respectiva con los indices de las etiquetas
    target1, target2: etiquetas de las entidades en cuestion
    '''
    # se consideran los indices para posteriormente colocarlos en el tsv 
    df_features = pd.DataFrame(columns=["docId","e1","start1","end1","e2","start2","end2", "features"])
    docs = text.keys()
    for d in docs:
        entities1 = df1[df1['docId'] == d]
        entities2 = df2[df2['docId'] == d]
        for i in entities1.index: 
            e1 = entities1.loc[i,'Entity']
            for j in entities2.index:
                e2 = entities2.loc[j,'Entity']
                start1 = entities1.loc[i,'Start']
                end1 = entities1.loc[i,'End']
                start2 = entities2.loc[j,'Start']
                end2 = entities2.loc[j,'End']
                features = features_entity1(e1, e2,text,d)
                df_features.loc[len(df_features)] = {"docId":d, "e1":e1,"start1": start1, "end1": end1, 
                                                      "e2":e2, "start2": start2, "end2": end2, "features": features}
    return df_features
```

```python
def generate_quantities(text, df1, df2, target1, target2):
    '''
    text: diccionario que contiene el id y el texto plano de cada documento
    df1, df2: DataFrame de la entidad respectiva con los indices de las etiquetas
    target1, target2: etiquetas de las entidades en cuestion
    '''
    # se consideran los indices para posteriormente colocarlos en el tsv 
    df_features = pd.DataFrame(columns=["docId","e1","start1","end1","e2","start2","end2", "features","annotSet"])
    docs = text.keys()
    for d in docs:
        entities1 = df1[df1['docId'] == d]
        entities2 = df2[df2['docId'] == d]
        count=1
        for i in entities1.index: 
            e1 = entities1.loc[i,'Entity']
            for j in entities2.index:
                e2 = entities2.loc[j,'Entity']
                start1 = entities1.loc[i,'Start']
                end1 = entities1.loc[i,'End']
                start2 = entities2.loc[j,'Start']
                end2 = entities2.loc[j,'End']
                features = features_entity1(e1, e2,text,d)
                df_features.loc[len(df_features)] = {"docId":d, "e1":e1,"start1": start1, "end1": end1, 
                                                      "e2":e2, "start2": start2, "end2": end2, "features": features,"annotSet":count}
            count+=1
    return df_features
```

```python
def tsv_quantities(docId, quantity_entity, quantity_unit, entity_property):
  '''
  '''
  df = pd.DataFrame(columns=["docId","annotSet","annotType","startOffset","endOffset","annotId","text","other"])
  quantities = quantity_entity[quantity_entity['docId'] == docId]
  properties = entity_property[entity_property['docId'] == docId]
  idx = quantities.index
  units = quantity_unit[quantity_unit['docId'] == docId]
  annot_set = quantities['annotSet'].unique()
  for i, q in enumerate(annot_set):

    text = quantities[quantities['annotSet'] == q]['e1'].values[0]
    # start: correspondiente al indice de inicio de la cantidad
    start = quantities[quantities['annotSet'] == q]['start1'].values[0]
    end = quantities[quantities['annotSet'] == q]['end1'].values[0]
    df.loc[len(df)] = {"docId": docId,"annotSet": q,"annotType": 'Quantity',"startOffset": start,"endOffset": end,"annotId": f'T{q}-1',"text": text}
    
    text2 = quantities[quantities['annotSet'] == q]['e2'].values[0]
    # start 2: correspondiente al indice de inicio de la entidad
    start2 = quantities[quantities['annotSet'] == q]['start2'].values[0]
    print(start2)
    end2 = quantities[quantities['annotSet'] == q]['end2'].values[0]

    # verificamos si tiene property, relacionando con el entity
    property = properties[properties['start1'] == start2]
    print(property)
    if len(property) > 0:
      property = property.values[0]
      # start 3: correspondiente al indice de inicio de la propiedad
      start3 = quantities[quantities['annotSet'] == q]['start2'].values[0]
      end3 = quantities[quantities['annotSet'] == q]['end2'].values[0]
      df.loc[len(df)] = {"docId": docId,"annotSet": q,"annotType": 'MeasuredProperty',"startOffset": start3,"endOffset": end3,"annotId": f'T{q}-2',
                         "text": property, "other": "{'HasQuantity':'T" + str(q) +"-1' }"}
      df.loc[len(df)] = {"docId": docId,"annotSet": q,"annotType": 'MeasuredEntity',"startOffset": start2,"endOffset": end2,"annotId": f'T{q}-3',
                       "text": text2, "other": "{'HasProperty': 'T" + str(q )+"-2' }"}
    else:
      df.loc[len(df)] = {"docId": docId,"annotSet": q,"annotType": 'MeasuredEntity',"startOffset": start2,"endOffset": end2,"annotId": f'T{q}-2',
                       "text": text2, "other": "{'HasQuantity':'T" + str(q) +"-1'}"}

  return df
```
